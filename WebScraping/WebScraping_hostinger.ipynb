```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "URL"
```

```python
httpRequest = requests.get(URL)
httpRequest
```

```python
httpRequest.headers
```

```python
httpRequest.text
```

```python
html = httpRequest.text
```

```python
parseHtml = BeautifulSoup(html, "html.parser")
parseHtml
```

```python
arrayOfprojectHtml = parseHtml.find_all("div" , {"class" :"h-carousel-card-wrapper__card"})
print(arrayOfprojectHtml)
```

```python
for price in arrayOfprojectHtml:
  try:
    print(price.find("div" , {"class" : "h-pricing-card__title t-h3"}).text)
    print(price.find("div", {"class" : "h-pricing-card__description t-body-3"}).text)
    print(price.find("span" , {"class" : "h-price__number t-header-currency"}).text)
    print(price.find("div" , {"class" : "h-pricing-card__features--grouped"}).text)
  except AttributeError:
    continue

```

```python
productPackageName = []
productItem_price = []
productPackage_detail = []
productDesc = []

for price in arrayOfprojectHtml:
  try:
    productPackageName.append(price.find("div" , {"class" : "h-pricing-card__title t-h3"}).text)
    productDesc.append(price.find("div" , {"class" : "h-pricing-card__description t-body-3"}).text)
    productItem_price.append(price.find("span" , {"class" : "h-price__number t-header-currency"}).text)
    productPackage_detail.append(price.find("div" , {"class" : "h-pricing-card__features--grouped"}).text)

  except AttributeError:
    continue
```

```python
p_Detail = [i.replace('\n','') for i in productPackage_detail]
p_Detail = [i.replace('\t','') for i in p_Detail]
p_Detail = [i.replace('\r','') for i in p_Detail]
p_price = [i.replace('\n','') for i in productItem_price]
p_Name = [i.replace('\n','') for i in productPackageName]
p_Desc = [i.replace('\n','') for i in productDesc]
```

```python
p_Name
```

```python
p_Desc = [i.replace('Ä°','i') for i in p_Desc]
p_Desc = [i.replace('Ã¶','ö') for i in p_Desc]
p_Desc = [i.replace('Ä±','ı') for i in p_Desc]
p_Desc = [i.replace('Ã¼','ü') for i in p_Desc]
p_Desc = [i.replace('Ã§','ç') for i in p_Desc]
p_Desc = [i.replace('Å\x9f','ş') for i in p_Desc]
```

```python
p_Desc
```

```python
p_price_TL = []
for i in p_price:
  p_price_TL.append(i + "TL /ay")
p_price_TL
```

```python
p_Detail = [i.replace('Ä°','İ') for i in p_Detail]
p_Detail = [i.replace('Ã\x96','Ö') for i in p_Detail]
p_Detail = [i.replace('Ä±','ı') for i in p_Detail]
p_Detail = [i.replace('Ã¼','ü') for i in p_Detail]
p_Detail = [i.replace('â\x82º','TL ') for i in p_Detail]
p_Detail = [i.replace('Å\x9f','ş') for i in p_Detail]
p_Detail = [i.replace('Ã\x9c','Ü') for i in p_Detail]
```

```python
p_Detail
```

```python
from datetime import datetime
import pytz
tz_Istanbul = pytz.timezone('Europe/Istanbul')

##Istanbul saati
current_dateTime = datetime.now(tz_Istanbul)
print("Istanbul:", current_dateTime.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
data = {'ProductName' : p_Name, 'ProductDesc' : p_Desc, 'productPrice' : p_price_TL,'ProductPackageDetail' : p_Detail, 'Date' : current_dateTime.strftime("%Y-%m-%d %H:%M:%S")}
```

```python
df = pd.DataFrame(data)
df
```

```python
df2 = df.to_json(orient ='index',date_format='iso')
print(df2)
```

```python
with open("sample2.json", "w") as outfile:
    outfile.write(df2)
```
